# Performance comparision between the models

```python
import pandas as pd
# Metrics from your Disease Grading (Single-task)
single_report = {
    "accuracy": 0.55,
    "macro_f1": 0.43,
    "weighted_f1": 0.54,
    "avg_precision": 0.50,
    "avg_recall": 0.43,
    "avg_dice": None  # Not applicable
}

# Metrics from your Multi-task Model
multi_report = {
    "accuracy": 0.78,
    "macro_f1": 0.83,
    "weighted_f1": 0.78,
    "avg_precision": 0.83,
    "avg_recall": 0.83,
    "avg_dice": 0.0818  # Binary dice across channels
}


df_compare = pd.DataFrame([single_report, multi_report], index=["Single-Task", "Multi-Task"])
display(df_compare.style.background_gradient(cmap='Blues'))
```

```python
import matplotlib.pyplot as plt

metrics_to_plot = ["accuracy", "macro_f1", "weighted_f1", "avg_precision", "avg_recall"]
df_compare[metrics_to_plot].plot(kind='bar', figsize=(10, 6), colormap='Set2')
plt.title("üìä Performance Comparison: Single-Task vs Multi-Task")
plt.ylabel("Score")
plt.ylim(0, 1)
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.legend(title="Metric")
plt.tight_layout()
plt.show()
```

## üß† Analysis: Single-Task vs Multi-Task

- ‚úÖ **Accuracy** improved from **55% ‚Üí 78%** in the multi-task model.
- ‚úÖ **Macro F1** jumped significantly, from **0.43 ‚Üí 0.83**, indicating better performance across all classes, not just the majority ones.
- üéØ **Weighted F1** also improved (0.54 ‚Üí 0.78), reflecting stronger handling of class imbalance.
- üé≠ **Multi-task model** additionally provides segmentation masks with an average Dice score of **0.0818** ‚Äî a first step toward lesion localization.
- ‚ö†Ô∏è **Single-task model** underperformed especially on minority classes (e.g., class 1 had precision and recall = 0).
